```python
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_s
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
```

```python
# Load the Diabetes dataset
columns = 'age sex bmi map tc ldl hdl tch ltg glu'.split() # Declare the columns names
diabetes = datasets.load_diabetes() # Call the diabetes dataset from sklearn
df = pd.DataFrame(diabetes.data, columns=columns) # load the dataset as a pandas data frame
y = diabetes.target # define the target variable (dependent variable) as y
```

```python
df.head()
```

```python
y[:5]
```

```python
# create training and testing vars
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

```python
# fit a model
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

```python
predictions[0:5]
```

```python
plt.scatter(y_test, predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
```

```python
print ('Test RMSE:', np.sqrt(mean_squared_error(predictions, y_test)))
print ('Train RMSE:', np.sqrt(mean_squared_error(lm.predict(X_train), y_train)))
```

# Cross Validation

```python
from sklearn.model_selection import KFold # import KFold
X = df  
y = diabetes.target  
kf = KFold(n_splits=3) # Define the split - into 2 folds 
kf.get_n_splits(X) # returns the number of splitting iterations in the cross-validator
```

```python
print(kf)
```

```python
# fit a model
lm = linear_model.LinearRegression()
kf.split(X)
```

```python
# fit a model
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
predictions.shape
```

```python
MSR=[]
for train_index, test_index in kf.split(X):
    print('TRAIN:', train_index, 'TEST:', test_index)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    model = lm.fit(X_train, y_train)
    predictions=lm.predict(X_test)
    MSR.append(np.sqrt(mean_squared_error(y_test,predictions)))
```

```python
sum(MSR)/len(MSR)
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn import metrics
```

```python
# Perform 6-fold cross validation
scores = cross_val_score(model, df, y, cv=2)
print ('Cross-validated scores:', scores)
```

```python
# Make cross validated predictions

predictions = cross_val_predict(model, df, y, cv=3)
plt.scatter(y, predictions)
print(predictions[0:5])
```

```python
np.sqrt(mean_squared_error(y,predictions))
```

```python
import pandas as pd
import numpy as np
from sklearn import datasets, linear_model
X = np.array([[5, 2], [5, 6], [1, 2], [3, 4]]) # create an array
y = np.array([1, 2, 3, 4]) # Create another array
```

```python
X=list(X)
y=list(y)
# fit a model
lm = linear_model.LinearRegression()
```

```python
num_folds = 2
subset_size =int(len(X)/num_folds)
train = []
test = []
cross_val={'Train': train, 'Test':test}
for i in range(num_folds):
    train.append(X[:i*subset_size] + X[(i+1)*subset_size:])
    test.append(X[i*subset_size:][:subset_size])
    train.append(y[:i*subset_size] + y[(i+1)*subset_size:])
    test.append(y[i*subset_size:][:subset_size])
```

```python
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler
  
# Read Data from CSV 
data = pd.read_csv('regression.csv') 
data.head() 
y = data['Value']
```

```python
data = data.drop(['Name', 'Nationality', 'Club', 'Wage', 'Position', 'Weight', 'Height', 'Contract Valid Until', 'Value'], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.2)
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
print ('Test RMSE:', np.sqrt(mean_squared_error(predictions, y_test)))
print ('Train RMSE:', np.sqrt(mean_squared_error(lm.predict(X_train), y_train)))
```

```python
# The coefficients
print('Coefficients: \n', lm.coef_)
# The mean squared error
print('Mean absolute error: %.2f'
      % mean_absolute_error(y_test, predictions ))

print('Mean squared error: %.2f'
      % mean_squared_error(y_test, predictions ))

print('Root mean squared error: %.2f'
      % np.sqrt(mean_squared_error(y_test, predictions )))

print('R-squared: %.2f'
      % r2_score(y_test, predictions ))
```

MinMaxScaler¶

```python
# Initialise the Scaler and Fit
norm = MinMaxScaler().fit(X_train)
```

```python
df_train=pd.DataFrame(columns = X_train.columns, data = norm.transform(X_train))
```

```python
df_test=pd.DataFrame(columns = X_test.columns, data = norm.transform(X_test))
```

```python
reg_MM = linear_model.LinearRegression()
model = reg_MM.fit(df_train, y_train)
predict_MM = reg_MM.predict(df_test)
print ('Test RMSE:', np.sqrt(mean_squared_error(predict_MM, y_test)))
print ('Train RMSE:', np.sqrt(mean_squared_error(reg_MM.predict(df_train), y_train)))
```

```python
# The coefficients
print('Coefficients: \n',reg_MM.coef_)
# The mean squared error
print('Mean absolute error: %.2f'
      % mean_absolute_error(y_test,predict_MM))

print('Mean squared error: %.2f'
      % mean_squared_error(y_test, predict_MM ))

print('Root mean squared error: %.2f'
      % np.sqrt(mean_squared_error(y_test, predict_MM )))

print('R-squared: %.2f'
      % r2_score(y_test, predict_MM ))
```
